```
import pandas as pd
```

```
sales=pd.read_csv('sales_file.csv')
```

```
sales.head()
```

```
sales['ordered_date'] = pd.to_datetime(sales['ordered_date'])
sales['ordered_date'] = pd.to_datetime(sales['ordered_date']).dt.date
```

```
sales=sales.drop(['delivered_date'],axis=1)
```

```
sales.head()
```

```
sales.isna().sum()
```

```
sales = sales.sort_values(['ordered_date','product_id'])
```

```
sales.head()
```

```
pnq=pd.read_csv('pnq.csv')
```

```
pnq.head()
```

```
pnq['date'] = pd.to_datetime(pnq['date']).dt.date
```

```
pnq.isna().sum()
```

```
pnq = pnq.sort_values(['product_id','date'])
```

```
pnq.head()
```

```
sales = sales.rename(columns = {'ordered_date': 'date'})
```

```
data=pd.merge(pnq,sales,on=['product_id','date'],how='left')
```

```
data.head()
```

```
data=data.fillna(0)
```

```
data.head()
```

```
data['Sales']=data['Qty_Stock']- data['Qty_Stock'].shift(-1)
```

```
data.head()
```

```
data.loc[data['Sales']<0]
```

```
data = data[data['Sales'] > 0]
```

```
data
```

```
import numpy as np
data['Target']=np.where(data['Oos Quantity']>0,1,0)
```

```
data.head()
```

```
x=data[['Normal_Selling','Qty_Stock','Disc_Rate','Ordered Quantity','Sales']]
y=data[['Target']]
```

```
data.loc[data['Target']==0]
```

```
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier
# fit model no training data
model = XGBClassifier()
model.fit(x,y.values.ravel())
y_pred = model.predict(x)
print("The Accuracy score is : ",accuracy_score(y,y_pred)*100,"%")
print(confusion_matrix(y,y_pred))
```
